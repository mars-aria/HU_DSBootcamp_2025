#### Missing Values Imputation Methods
Missing value imputation is a technique used to replace missing values in a dataset with estimated values. This is done to preserve the data and make it more complete for analysis. There are many different imputation techniques available, each with its own advantages and disadvantages.

Some of the most common imputation techniques include:

- Mean imputation: This is the simplest imputation technique. It involves replacing missing values with the mean of the observed values for that variable. This is a quick and easy technique, but it can introduce bias if the distribution of the missing values is not normal. Mean as a measure is greatly affected by outliers or if the distribution of the data or column is not normally-distributed. Therefore, it’s wise to first check the distribution of the column before deciding if to use a mean imputation or median imputation.
- Median imputation: This is similar to mean imputation, but it replaces missing values with the median of the observed values for that variable. This is a more robust technique than mean imputation, but it can still introduce bias if the distribution of the missing values is not symmetric.
- Mode imputation: This replaces missing values with the most frequent value of the observed values for that variable. This is a simple technique that is not as likely to introduce bias as mean or median imputation. However, it can be problematic if the most frequent value is not representative of the overall distribution of the data.
- Regression imputation: This technique uses a regression model to predict the missing values. The model is trained on the observed values for the variable with missing values and other related variables. The predicted values are then used to replace the missing values. This is a more sophisticated technique than the other imputation techniques, but it can be more time-consuming and computationally expensive.

The best imputation technique to use depends on the specific dataset and the analysis that you are trying to do. In general, it is a good idea to try multiple imputation techniques and see which one works best for your data.

- Mean imputation works better if the distribution is normally-distributed or has a Gaussian distribution, while median imputation is preferable for skewed distribution(be it right or left)
- If the values are not normally distributed and show some skewness toward extreme high values or extreme low values consider using the median for imputation. The median is not suceptible to skewness in the data.
- Sometimes it is advisable to drop the columns when the NULL values are significantly more than the other values present.

### Research Question: Which political party has most influenced the banning of more books in the U.S.?
This visualization technique describes the initial research question by showcasing the distribution of books per state that were banned between the years 2021 and 2022. It also utilizes color to illustrate the political affiliation of each censorship state in the U.S. at the time. In this context, a top censorship state is classified as a state that has had more than one occurrence of book banning appear in the dataset.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# List of states
bars = ["Texas", "Florida", "Pennsylvania", "Tennessee", "Oklahoma",
        "Michigan", "Kansas", "Wisconsin", "Missouri", "Idaho",
        "Georgia", "Mississippi", "Virginia", "Indiana", "North Carolina",
        "New York", "Utah", "Iowa", "Illinois", "South Dakota",
        "Washington", "Ohio", "New Jersey"]

# Number of observations
bar_heights = [801, 566, 458, 349, 43, 41, 30, 29, 27, 26,
               23, 22, 19, 18, 16, 13, 12, 8, 5, 5, 5, 4, 3]

# List of colors you pick directly
# Use 'red' for Republican and 'blue' for Democrat
bar_colors = ["red", "red", "blue", "red", "red",
              "blue", "red", "blue", "red", "red",
              "blue", "red", "blue", "red", "red",
              "blue", "red", "red", "blue", "red",
              "blue", "red", "blue"]

# Plot
plt.figure(figsize=(14, 6))
x_pos = np.arange(len(bars))
plt.bar(x_pos, bar_heights, color=bar_colors)

# Labels and title
plt.title('Political Affiliations of Top Book-Banning U.S. States [2021-2022]', fontweight='bold', fontsize=18)
plt.xlabel('State', fontweight='bold', fontsize=15)
plt.ylabel('Total Number of Bans', fontweight='bold', fontsize=15)

# X-ticks
plt.xticks(x_pos, bars, rotation=45, fontsize=10)

# Legend
red_patch = mpatches.Patch(color='red', label='Republican')
blue_patch = mpatches.Patch(color='blue', label='Democrat')
plt.legend(handles=[red_patch, blue_patch], title='Party', loc='upper right')

plt.show()
```

### Summary of the Findings

  - The top 5 censorship states were Texas, Florida, Pennsylvania, Tennessee, and Oklahoma.

### Research Question: Which type of ban has been most prevalent in each state within the last few years?
This visualization technique describes the secondary research question by showcasing the distribution of ban types per state between the years 2021 and 2022. It also utilizes color to illustrate the difference in ban type for each censorship state in the U.S.

```python
# Splits the 'Type of Ban' column into multiple rows
df_exploded = DataFrame2.copy()
df_exploded['Type of Ban'] = df_exploded['Type of Ban'].str.split(', ')
df_exploded = df_exploded.explode('Type of Ban')
```

```python
# Renames two of the ban types for the sake of simplicity
DataFrame2['Type of Ban'] = DataFrame2['Type of Ban'].replace({
    'Banned in Libraries and Classrooms': 'Banned in Both',
    'Banned Pending Investigation': 'Pending Book Bans'
})
```

```python
# Groups by State and Type of Ban
totals_by_state = DataFrame2.groupby(['State', 'Type of Ban']).size().reset_index(name='Total Bans')

# Pivots data for stacked bars
pivot_df = totals_by_state.pivot(index='State', columns='Type of Ban', values='Total Bans').fillna(0)

# Sorts states by total bans (for nicer visualization)
pivot_df['Total'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values('Total', ascending=False).drop(columns='Total')

# Plots the data as a stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(16, 8), colormap='tab20')

plt.title('Frequency Distribution of Book Ban Types in U.S. States [2021-2022]', fontweight='bold', fontsize=20)
plt.xlabel('State', fontweight='bold', fontsize=18)
plt.ylabel('Total Number of Bans', fontweight='bold', fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=16)
plt.legend(title='Type of Ban', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Summary of the Findings

#### Observations: Stacked Bar Chart
From this visualization, we observe the following:
  - The censorship state with the most books banned in libraries: Texas
  - The censorship state with the most books banned in classrooms: Pennsylvania
  - The censorship state with the most books banned in both libraries and classrooms: Florida
  - The censorship state with the most pending book ban case investigations: Texas
